```python
def getMask(img,background,threshold, is_black = None):
    diff = cv2.absdiff(img,background)
    diff = diff.mean(axis=2)
    mask = diff > threshold
    mask = mask[:,:,np.newaxis]
    if(is_black is not None):
        mask_b =   is_black & (diff > threshold - 20)
        mask_b = mask_b[:,:,np.newaxis] | mask
        mask = mask_b
    return mask
```

```python
from numba import jit
@jit(target_backend = 'cuda')
def getMask_gpu(img,background,threshold, is_black = None):
    diff = cv2.absdiff(img, background)
    diff = diff.mean(axis=2)
    mask = diff > threshold
    mask = mask[:,:,np.newaxis]
    if(is_black is not None):
        mask_b =   is_black & (diff > threshold - 20)
        mask_b = mask_b[:,:,np.newaxis] | mask
        mask = mask_b
    return mask
```

```python
def getCentroid(frame, mask, plot = True):
    centroid = np.empty((0,2))
    if plot:
        plt.imshow(frame)
    objects = frame*mask.astype('uint8')
    #DBSCAN을 이용하여 개체별로 clustering
    nonzero_index = np.transpose(np.nonzero(objects))
    features = nonzero_index
    clustering = DBSCAN(eps = 2, min_samples=10).fit(features)
    
    core_samples_mask = np.zeros_like(clustering.labels_, dtype=bool)
    core_samples_mask[clustering.core_sample_indices_] = True
    labels = clustering.labels_
    unique_labels = set(labels)
    colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]
    for i,k in enumerate(unique_labels):
        if k == -1: #노이즈
            break
        class_member_mask = (labels == k)
        xy = nonzero_index[class_member_mask & core_samples_mask]
        x = xy[:,1].mean()
        y = xy[:,0].mean()
        centroid = np.append(centroid,np.array([[x,y]]),axis=0)
        if plot:
            plt.plot(xy[:, 1], xy[:, 0],"o",markerfacecolor=tuple(colors[i]),markeredgecolor="None", markersize = 1, alpha = 1)

            plt.plot(x, y, 'o', markerfacecolor='blue', markeredgecolor="None", markersize = 5, alpha = 1)
            plt.text(x, y, i, fontsize=20)

    return centroid
```
